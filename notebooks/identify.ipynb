## 2. Decompose the samples
Now, we can use the `mutagene signature` command to process the MAF file and see the results.

- The option `-i` specifies the input file
- `-g` specifies the reference genome
- `-s` specifies the signature set
    - `5`: MutaGene 5
    - `10`: MutaGene 10
    - `30`: COSMIC v2
    - `49`: COSMIC v3
- `-f` specifies the input file format

````python
!mutagene signature -i ./skcm_yale/data_mutations_mskcc.txt -g hg19.2bit -s 10 -f MAF
````

The output displays all the samples and their respective decompositions. For example:
